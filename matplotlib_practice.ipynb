```python
# Import dependencies

%matplotlib inline

import matplotlib.pyplot as plt


```

```python
# Set the x-axis t a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to the a list of foloats as the total fare in US dollars accumulated for each month.

y_axis = [10.2, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot

plt.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.

plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

# Create a title.

plt.title("PyBer Fare by Month")

# Add the legend.

plt.legend()
```
